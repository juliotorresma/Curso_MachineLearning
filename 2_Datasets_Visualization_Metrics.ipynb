## Tipos de funciones

#### Clasificaci√≥n

- **Num√©rico**: N√∫meros, pueden ser enteros o continuos. Por ejemplo: edad, precio, temperatura.
- **Ordinal**: Los valores de estas caracter√≠sticas no son n√∫meros, pero sus valores tienen un orden espec√≠fico. Por ejemplo: ¬øC√≥mo te sientes hoy? Muy infeliz, infeliz, neutral, feliz o muy feliz. O acad√©mico: escuela primaria, secundaria, universidad, maestr√≠a, doctorado.
- **Categ√≥rico o nominal**: Los valores de estas caracter√≠sticas representan categor√≠as. No se pueden ordenar. Por ejemplo, g√©nero, color de cabello, lugar de nacimiento.


## <font color = 'blue'> Funciones num√©ricas </font>

<u> Algunas m√©tricas que pueden resultar √∫tiles para comprender las funciones num√©ricas son: </u>

- M√≠nimo
- M√°ximo
- Rango: $ M√°ximo - m√≠nimo $
- Promedio o media: $ \ bar {x} = \ frac {\ sum_i x_i} {n} $
- Mediana: si ordenamos las muestras, la mediana es el valor que se encuentra justo en el medio.
- Varianza: poblaci√≥n $ \ sigma ^ 2 = \ frac {\ sum_i (x_i- \ bar {x}) ^ 2} {n} $ muestra $ s ^ 2 = \ frac {\ sum_i (x_i- \ bar {x} ) ^ 2} {n-1} $
- Desviaci√≥n est√°ndar: poblaci√≥n $ \ sigma = \ sqrt {\ sigma ^ 2} $ muestra $ s = \ sqrt {s ^ 2} $. La desviaci√≥n est√°ndar puede verse como el promedio de todas las variaciones entre los datos y el promedio.

- Cuartiles: divide el n√∫mero de puntos de datos en cuatro partes. Los datos deben ordenarse de menor a mayor. El valor del 25% de los datos est√° por debajo del primer cuartil (Q1). El segundo cuartil (Q2) es la mediana de un conjunto de datos; por tanto, el 50% de los datos se encuentra por debajo de este punto. El valor del 75% de los datos est√° por debajo del primer cuartil (Q3).

- Rango intercuartil: es √∫til para encontrar valores at√≠picos. $ IQR = Q_3- Q_1 $. Un valor at√≠pico es una muestra cuyo valor $ x $ es $ x <Q_1-1.5 * IQR $ o $ x> Q_3 + 1.5 * IQR $.





<u> Para medir la dependencia entre dos caracter√≠sticas num√©ricas, podemos usar la covarianza y la correlaci√≥n. </u>

**_Covarianza_**

Es un valor num√©rico que mide la dependencia lineal entre dos caracter√≠sticas num√©ricas.

Poblaci√≥n $ \ sigma_ {xy} = \ frac {\ sum_i (x_i- \ bar {x}) (y_i- \ bar {y})} {n} $ muestra $ \ sigma_ {xy} = \ frac {\ sum_i ( x_i- \ bar {x}) (y_i- \ bar {y})} {n-1} $

Interpretaci√≥n:

- Un valor positivo indica que un aumento en una variable da como resultado un aumento en la otra variable.
- Un valor negativo indica que un aumento en una variable da como resultado el cambio opuesto en la otra variable.
- Un valor cercano a cero indica que no hay dependencia lineal. Sin embargo, podr√≠a haber una dependencia no lineal.

[Ver imagen] (https://en.wikipedia.org/wiki/Covariance#/media/File:Covariance_trends.svg)

El problema de la covarianza es que no tiene un rango limitado. Es dif√≠cil saber si un valor de covarianza de 343 representa una dependencia fuerte o d√©bil.

**_Correlaci√≥n_**

El coeficiente de correlaci√≥n de Pearson, com√∫nmente llamado coeficiente de correlaci√≥n, es un valor num√©rico que mide la relaci√≥n lineal entre dos caracter√≠sticas num√©ricas. La diferencia con la covarianza es que tiene un rango limitado entre -1 y 1.

Poblaci√≥n $ r = \ frac {\ sigma_ {xy}} {\ sigma_x \ sigma_y} = \ frac {Cov (x, y)} {StdDev (x) StdDev (y)} $

Muestra $ r = \ frac {\ sum_i (x_i- \ bar {x}) (y_i- \ bar {y})} {\ sqrt {\ sum_i (x_i - \ bar {x}) ^ 2} \ sqrt {\ sum_i (y_i - \ bar {y}) ^ 2}} $

Interpretaci√≥n:

- El signo representa lo mismo que la covarianza.
- Es un n√∫mero entre -1 y 1
    - Cuanto m√°s cerca de -1 o 1, mayor ser√° la dependencia lineal.
    - Los valores cercanos a 0 indican que no hay dependencia lineal.

[Ver imagen] (https://en.wikipedia.org/wiki/Correlation_and_dependence#/media/File:Correlation_examples2.svg)

Desde mi perspectiva personal, es mucho mejor utilizar la correlaci√≥n que la covarianza para medir la dependencia lineal entre caracter√≠sticas num√©ricas.

**_Entrop√≠a_**

La entrop√≠a, tambi√©n llamada entrop√≠a de la informaci√≥n o entrop√≠a de Shannon, mide la **sorpresa** o el **caos** en los valores de una variable. Formalmente, mide la uniformidad de la variable. Se puede calcular de la siguiente manera:

$ H (X) = - \ sum_x P (x) log_2 P (x) $
